## Investor Age Group Distribution

```python
age_data = transactions["age_group"].value_counts()

plt.figure(figsize=(7,7))

plt.pie(
    age_data,
    labels=age_data.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Investor Distribution by Age Group")

plt.savefig("../reports/charts/age_group_distribution.png", dpi=300)

plt.show()
```

**Insight**

The 26–45 age groups account for the majority of investors, indicating that working professionals contribute most to mutual fund investments.

## SIP Investment Amount by Age Group

```python
sip_data = transactions[
    transactions["transaction_type"] == "SIP"
]

plt.figure(figsize=(9,6))

sns.boxplot(
    data=sip_data,
    x="age_group",
    y="amount_inr"
)

plt.title("SIP Amount Distribution by Age Group")

plt.xlabel("Age Group")

plt.ylabel("Investment Amount (₹)")

plt.tight_layout()

plt.savefig("../reports/charts/sip_boxplot_age.png", dpi=300)

plt.show()
```

**Insight**

SIP investment amounts vary across age groups, with middle-aged investors generally contributing larger and more consistent investment amounts.

## Gender Distribution

```python
gender = transactions["gender"].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    gender,
    labels=gender.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Investor Gender Distribution")

plt.savefig("../reports/charts/gender_distribution.png", dpi=300)

plt.show()
```

**Insight**

The investor base is distributed across both genders, providing a balanced participation in mutual fund investments.

## State-wise SIP Investment

```python
state_sip = (
    transactions[transactions["transaction_type"] == "SIP"]
    .groupby("state")["amount_inr"]
    .sum()
    .sort_values(ascending=True)
)

plt.figure(figsize=(10,8))

state_sip.plot(kind="barh")

plt.title("Total SIP Investment by State")
plt.xlabel("Investment Amount (₹)")
plt.ylabel("State")

plt.tight_layout()

plt.savefig("../reports/charts/state_sip.png", dpi=300)

plt.show()
```

**Insight**

States with larger urban investor populations contribute the highest SIP investments, highlighting stronger mutual fund penetration in financially active regions.

## T30 vs B30 Investor Distribution

```python
tier = transactions["city_tier"].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    tier,
    labels=tier.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Investor Distribution by City Tier")

plt.savefig("../reports/charts/city_tier.png", dpi=300)

plt.show()
```

**Insight**

T30 cities account for the majority of investors, while B30 cities continue to contribute a significant share, indicating expanding mutual fund adoption beyond metro areas.

## Folio Count Growth (2022–2025)

```python
industry["month"] = pd.to_datetime(industry["month"])

plt.figure(figsize=(12,6))

plt.plot(
    industry["month"],
    industry["total_folios_crore"],
    marker="o",
    linewidth=2
)

plt.title("Total Mutual Fund Folios Growth (2022–2025)")
plt.xlabel("Month")
plt.ylabel("Folios (Crore)")

plt.annotate(
    "13.26 Cr",
    xy=(industry["month"].iloc[0], industry["total_folios_crore"].iloc[0]),
    xytext=(industry["month"].iloc[3], industry["total_folios_crore"].iloc[0]+0.4),
    arrowprops=dict(arrowstyle="->")
)

plt.annotate(
    "26.12 Cr",
    xy=(industry["month"].iloc[-1], industry["total_folios_crore"].iloc[-1]),
    xytext=(industry["month"].iloc[-8], industry["total_folios_crore"].iloc[-1]-0.8),
    arrowprops=dict(arrowstyle="->")
)

plt.grid(alpha=0.3)

plt.tight_layout()

plt.savefig("../reports/charts/folio_growth.png", dpi=300)

plt.show()
```

**Insight**

Mutual fund folios nearly doubled during the period, reflecting strong retail participation and increasing investor confidence.

## Fund Return Correlation Matrix

```python
corr = scheme[
    [
        "return_1yr_pct",
        "return_3yr_pct",
        "return_5yr_pct",
        "alpha",
        "beta",
        "sharpe_ratio",
        "sortino_ratio",
        "std_dev_ann_pct",
        "max_drawdown_pct"
    ]
].corr()

plt.figure(figsize=(10,8))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Matrix of Performance Metrics")

plt.tight_layout()

plt.savefig("../reports/charts/correlation_matrix.png", dpi=300)

plt.show()
```

**Insight**

Long-term returns show a positive relationship with risk-adjusted performance metrics, while higher volatility and drawdowns negatively influence fund performance.

## Portfolio Sector Allocation

```python
sector = (
    portfolio
    .groupby("sector")["weight_pct"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,8))

plt.pie(
    sector,
    labels=sector.index,
    autopct="%1.1f%%",
    startangle=90,
    wedgeprops=dict(width=0.4)
)

plt.title("Portfolio Sector Allocation")

plt.savefig("../reports/charts/sector_donut.png", dpi=300)

plt.show()
```

**Insight**

The portfolio is diversified across sectors, with a few major sectors contributing the highest allocation while maintaining balanced exposure across the market.
